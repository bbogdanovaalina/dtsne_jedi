```python
X_plot, y_plot = get_gaussian_data(2, 7, 300)

Y_plot = dtsne(X_plot, n_iter=100, num_processes=16)
def plot_2(X, Y, y):
    plt.figure(figsize=(15,8))
    plt.subplot(1,2,1)
    plt.title('Gaussian data with 7 clusters in 2 dimensional space, number of samples=2000')
    sc = plt.scatter(X[:, 0], X[:, 1], 3, y)
    plt.legend(*sc.legend_elements(), title="Clusters")
    plt.subplot(1,2,2)
    plt.title('dtSNE algorithm, number of samples=2000, n_iter=100')
    sc = plt.scatter(Y[:, 0], Y[:, 1], 3, y)
    plt.legend(*sc.legend_elements(), title="Clusters")
    plt.show()

plot_2(X_plot, Y_plot, y_plot) 
```

```python
X, y, n_samples = get_gaussian_data_jedi(n_clusters=7)
sum_X = np.sum(np.square(X), 1)
D = np.add(np.add(-2 * np.dot(X, X.T), sum_X).T, sum_X)
Z = np.random.randn(X.shape[0], X.shape[0]) ** 2 
Z = Z + Z.T
Z[0:n_samples[0] + n_samples[1], 0:n_samples[0] + n_samples[1]] = D[0:n_samples[0] + n_samples[1], 0:n_samples[0] + n_samples[1]]

alpha=1.0
beta=0.8
n_components=2
initial_dims=50
perplexity=50.0
n_iter = 100
Y = jedi(X, Z, alpha, beta, n_components, perplexity, n_iter, 0, verbose=1)
```

```python
plt.figure(figsize=(15,8))
plt.subplot(1,2,1)
plt.title('Gaussian data with 7 clusters in 2 dimensional space, number of samples=2000')
sc = plt.scatter(X[:, 0], X[:, 1], 3, y)
plt.legend(*sc.legend_elements(), title="Clusters")
plt.subplot(1,2,2)
plt.title('JEDI algorithm, number of samples=2000, n_iter=100')
sc = plt.scatter(Y[:, 0], Y[:, 1], 3, y)
plt.legend(*sc.legend_elements(), title="Clusters")
plt.show()
```
